```python
import numpy as np
import pandas as pd
import seaborn as sns

from ISLP import load_data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

df = load_data("Weekly")

%matplotlib inline
```

```python
df
```

```python
df['Dir'] = (df['Direction'] == "Up").astype(int)
```

```python
lr = LogisticRegression(C=10**5)
X = df[['Lag1','Lag2']]
y = df['Dir']
mod = lr.fit(X,y)
mod.coef_
```

```python
X_one = df[['Lag1','Lag2']][1:]
y_one = df['Dir'][1:]
mod = lr.fit(X_one,y_one)
mod.coef_
```

```python
mod.predict([X.iloc[0]]),y[0]
# This observation was not correctly classified
```

```python
n = len(X)
errors = np.zeros(n)

for i in range(n):
    one_out  = ~X.index.isin([i])
    # i.
    mod.fit(X[one_out], y[one_out])
    # ii. iii. iv.
    if mod.predict([X.iloc[i]]) != y[i]:
        errors[i] = 1
```

```python
errors.mean()
```
