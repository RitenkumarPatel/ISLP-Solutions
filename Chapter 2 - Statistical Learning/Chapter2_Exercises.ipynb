```python
auto = pd.read_csv("Auto.csv",
                   na_values=["?"])
auto = auto.dropna()
auto.count()
```

```python
auto.head()
```

```python
# Quantitative: mpg, displacement, horsepower, weight, acceleration, year
# Qualitative: cylinders, origin, name
# Ranges:
quantitative = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration', 'year']
def printer(auto):
    for item in quantitative:
        print("Range of", item, ":", auto[item].max() - auto[item].min())
        print("Mean of", item, ":", auto[item].mean())
        print("Standard Deviation of", item, ":", auto[item].std())
        print("\n")
printer(auto)
```

```python
auto_sliced = auto.drop(auto.index[9:85])
printer(auto_sliced)
```

```python
pd.plotting.scatter_matrix(auto[['weight','horsepower','acceleration','displacement','mpg']]);
```

```python
# Positively correlated: horsepower vs weight, displacement vs horsepower, displacement vs weight
# Unrelated: acceleration vs weight, year and basically anything

# Horsepower and weight look like good indicators of mpg, displacement has a weaker correlation as well
```
